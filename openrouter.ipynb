```python
import requests

# 🔐 Use NEW API key (no < >)
API_KEY = "API"

url = "https://openrouter.ai/api/v1/chat/completions"

headers = {
    "Authorization": f"Bearer {API_KEY}",
    "Content-Type": "application/json",
}

# -----------------------------
# 🔹 First API Call
# -----------------------------
payload1 = {
    "model": "moonshotai/kimi-k2-thinking",
    "messages": [
        {
            "role": "user",
            "content": "How many r's are in the word 'strawberry'?"
        }
    ],
    "reasoning": {"enabled": True},
    "max_tokens": 800   # 💰 Control cost
}

res1 = requests.post(url, headers=headers, json=payload1)

# 🛡️ Safe parsing
try:
    data1 = res1.json()
except:
    print("❌ Failed to parse first response")
    print(res1.text)
    exit()

# ❌ Handle API errors
if "error" in data1:
    print("❌ API Error (First Call):", data1["error"])
    exit()

# ✅ Extract assistant message
assistant_msg = data1["choices"][0]["message"]

# -----------------------------
# 🔹 Preserve conversation
# -----------------------------
messages = [
    {"role": "user", "content": "How many r's are in the word 'strawberry'?"},
    {
        "role": "assistant",
        "content": assistant_msg.get("content"),
        "reasoning_details": assistant_msg.get("reasoning_details")
    },
    {"role": "user", "content": "Are you sure? Think carefully."}
]

# -----------------------------
# 🔹 Second API Call
# -----------------------------
payload2 = {
    "model": "moonshotai/kimi-k2-thinking",
    "messages": messages,
    "reasoning": {"enabled": True},
    "max_tokens": 800   # 💰 Control cost again
}

res2 = requests.post(url, headers=headers, json=payload2)

# 🛡️ Safe parsing
try:
    data2 = res2.json()
except:
    print("❌ Failed to parse second response")
    print(res2.text)
    exit()

# ❌ Handle API errors
if "error" in data2:
    print("❌ API Error (Second Call):", data2["error"])
    exit()

# ✅ Extract clean answer
try:
    final_answer = data2["choices"][0]["message"]["content"]

    print("\n💡 Final Answer:\n")
    print(final_answer)

except:
    print("⚠️ Unexpected response format:")
    print(data2)
```
